```python
# Arithmetic Operators
a = int(input())
b = int(input())
print("Addition of a and b:",a+b)
print("Substraction of a and b:",a-b)
print("Multiplication of a and b:",a*b)
print("Division of a and b:",a/b)
print("Floor Division of a and b:",a//b)
print("Modulus of a and b:",a%b)
print("Power of a and b:",a**b)
```

```python
# Comparision Operator
print(a==b)
print(a!=b)
print(a>b)
print(a<b)
print(a<=b)
print(a>=b)
```

```python
# Logical Operators
print(a>30 and a<10)
print(a>30 or a<10)
print(not(a>18))
```

```python
# Assignment Operators
a +=10
a -= 5
a *= 3
a //= 2
print(a)
```

```python
# Excercise 2 : Write a program that asks for your age and checks:
#If age < 18 → print "Minor"
#If age between 18 and 60 → print "Adult"
#If age > 60 → print "Senior Citizen"

age = int(input("Enter your age: "))

if age < 18:
    print("Minor")
elif age <= 60:   # 18 to 60
    print("Adult")
else:
    print("Senior Citizen")
```
